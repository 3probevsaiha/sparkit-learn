```python
import numpy as np
import scipy.sparse as sp

from sklearn.utils.testing import assert_array_equal, assert_equal
from sklearn.utils.testing import assert_array_almost_equal, assert_raises

from sklearn.datasets import make_classification
from sklearn.naive_bayes import MultinomialNB

from spylearn import rdd

from spylearn.naive_bayes import SparkMultinomialNB
```

```python
def generate(size, blocks):
    X, y = make_classification(n_classes=4,
                               n_samples=size, n_features=20,
                               n_informative=10, n_redundant=0,
                               n_clusters_per_class=1,
                               random_state=42)
    X, y = np.abs(X), y

    X_rdd = sc.parallelize(X)
    y_rdd = sc.parallelize(y)
    Z = rdd.block(X_rdd.zip(y_rdd), size / blocks if blocks is not None else None)  # RDD
    
    return X, y, Z
```

```python
def run(size, blocks):
    X, y, Z = generate(size, blocks)
    local = MultinomialNB()
    dist = SparkMultinomialNB()

    y_local = local.fit(X, y)#.predict(self.X)
    y_dist = dist.fit(Z, classes=np.unique(y))#.predict(self.Z.column(0))

    coef_error = np.mean(np.abs(y_local.class_log_prior_ - y_dist.class_log_prior_))
    intercept_error = np.mean(np.abs(y_local.feature_log_prob_ - y_dist.feature_log_prob_))

    return coef_error, intercept_error
    # assert_array_equal(y_local.class_log_prior_, y_dist.class_log_prior_)
    # assert_array_almost_equal(y_local.feature_log_prob_, y_dist.feature_log_prob_)
    # assert_array_equal(y_local, np.concatenate(y_dist.collect()))
```

```python
results = []
for i in np.logspace(3, 6, 8):
    for j in [4, 8, 16, 32]:
        results.append((i, j) + run(int(i), j))
```
